```python
import pandas as pd
```

```python
df = pd.read_csv("data/raw/1112inmigall.csv")
```

```python
df.columns
```

```python
fips_dict = df[["statefips","state","state_name"]].drop_duplicates(subset = ["statefips"])
```

```python
region_map = {
    "Northeast": ["CT","ME","MA","NH","NJ","NY","PA","RI","VT"],
    "Midwest":   ["IL","IN","IA","KS","MI","MN","MO","NE","ND","OH","SD","WI"],
    "South":     ["AL","AR","DE","DC","FL","GA","KY","LA","MD","MS","NC","OK","SC","TN","TX","VA","WV"],
    "West":      ["AK","AZ","CA","CO","HI","ID","MT","NV","NM","OR","UT","WA","WY"]
}

division_map = {
    "New England":        ["CT","ME","MA","NH","RI","VT"],
    "Middle Atlantic":    ["NJ","NY","PA"],
    "East North Central": ["IL","IN","MI","OH","WI"],
    "West North Central": ["IA","KS","MN","MO","NE","ND","SD"],
    "South Atlantic":     ["DE","DC","FL","GA","MD","NC","SC","VA","WV"],
    "East South Central": ["AL","KY","MS","TN"],
    "West South Central": ["AR","LA","OK","TX"],
    "Mountain":           ["AZ","CO","ID","MT","NV","NM","UT","WY"],
    "Pacific":            ["AK","CA","HI","OR","WA"]
}

# Convert the region dict into state → region mapping
state_to_region = {
    state: region
    for region, states in region_map.items()
    for state in states
}

# Convert the division dict into state → division mapping
state_to_division = {
    state: division
    for division, states in division_map.items()
    for state in states
}

fips_dict["region"] = fips_dict["state"].map(state_to_region)
fips_dict["division"] = fips_dict["state"].map(state_to_division)
```
